**2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.**

```python
#check which two days are busiest and slowest
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

The busiest days are Saturday(0) & Sunday(1).  The slowest days are Tuesday(3) & Wednesday(4).

```python
#create result2 variable using for and if statements highligting two busiest and two slowest days
result2 = []

for value in ords_prods_merge['orders_day_of_week']:
  if value == 0 or value == 1:
    result2.append('Busiest days')
  elif value == 4 or value == 3:
    result2.append('Slowest days')
  else:
    result2.append('Regular Days')
```

**3. Check the values of this new column for accuracy. Note any observations in markdown format.**

```python
#create new column titled 'busiest_days' from result2
ords_prods_merge['busiest_days'] = result2
```

```python
#check frequency
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
#check for accuracy
value_counts = ords_prods_merge[['orders_day_of_week', 'busiest_days']].value_counts(dropna = False)
print (value_counts)
```

As can be seen from above, the frequency counts match for all the variables in the two columns: orders_day_of_week and busiest_days.  In addition, 0 and 1 correspond with Busiest days.  Also, 3 and 4 correspond with Slowest days. So the code was executed successfully.

**4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”**

```python
#check frequency counts of order_hour_of_day
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

I will divide the hours by three time periods.  With 'Most orders' being hours between 9(9am) and 16(4pm). 'Average orders' will be between 17(5pm) and 22(10pm) and also between 7(7am) and 8(8am). 'Fewest orders' will be between 23(11pm) and 0 (12am) and also between 1(1am) and 6(6am).  

```python
#create variable with order_hour_of_day divided into 'Most orders', 'Average Orders', and 'Fewest Orders'
result3 = []

for value in ords_prods_merge['order_hour_of_day']:
  if value in [9, 10, 11, 12, 13, 14, 15, 16]:
    result3.append('Most orders')
  elif value in [0, 1, 2, 3, 4, 5, 6, 23]:
    result3.append('Fewest orders')
  else:
    result3.append('Average orders')
   
```

```python
#create new column with results3 called busiest_period_of_day
ords_prods_merge['busiest_period_of_day'] = result3
```

**5. Print the frequency for this new column.**

```python
#check frequency counts of busiest_period_of_day
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```

```python
#check accuracy
value_counts1 = ords_prods_merge[['order_hour_of_day', 'busiest_period_of_day']].value_counts(dropna = False)
print (value_counts1)
```

Everything matches as it should, order_hour_of_day values 9-16 all have 'Most orders', values 7-8 and 17-22 have 'Average orders', and 0-6 and 23 have 'Fewest orders'.
