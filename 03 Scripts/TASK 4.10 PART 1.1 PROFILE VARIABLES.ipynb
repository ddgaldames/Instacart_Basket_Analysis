**3. Regional based variable**

```python
#check to make sure the state names are unifrom 

cust_ords_prods['state'].value_counts(dropna = False).sort_index()
```

State names are uniform throughout, so we can go ahead and create the new columns

```python
#create for loop with the states placed into regions

region = []
for value in cust_ords_prods['state']:
    if value in ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']:
        region.append('Northeast')
    elif value in ['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']:
        region.append('Midwest')
    elif value in ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']:
        region.append('South')
    else:
        region.append('West')
```

```python
#create new column in cust_ords_prods with regults from region for loop

cust_ords_prods['region'] = region
```

```python
#check for loop was successful with frequency count of each region

cust_ords_prods['region'].value_counts(dropna = False)
```

```python
10791885+8292913+7597325+5722736
```

```python
#check if region column has same number of rows as data set

len(cust_ords_prods['region'])
```

From the two checks, we can ascertain that the for loop code was executed succesfully.

As can be seen from the above bar chart, all the regions have significantly more low spenders than high spenders. 

**4. Customer_activity flag**

```python
#create for loop with new column customer_activity to check when max_orders are less then 5

customer_activity = []
for value in cust_ords_prods['max_order']:
    if value < 5:
        customer_activity.append('low-activity')
    else:
        customer_activity.append('high-activity')
```

```python
#add new column customer_activity to cust_ords_prods

cust_ords_prods['customer_activity'] = customer_activity
```

```python
#check number of rows of customer_activity to ensure it is the same as cust_ords_prods

len(cust_ords_prods['customer_activity'])
```

```python
#check frequency counts of cust_ords_prods

cust_ords_prods['customer_activity'].value_counts(dropna = False)
```

```python
30964564+1440295
```

So, we have 30,965,564 rows with high_activity, so when we drop rows with low_activity.  This is the number of rows the new data set should have.

```python
#drop rows with low_activity and keep high_activity in a new data set called high_act_cust_ords_prods

high_act_cust_ords_prods = cust_ords_prods[cust_ords_prods['customer_activity'] == 'high-activity']
```

```python
#check number of rows of high_act_cust_ords_prods without low_activity

len(high_act_cust_ords_prods['customer_activity'])
```

The number matches the number of high_activity flags from line out[18]  So the code was executed perfectly.
